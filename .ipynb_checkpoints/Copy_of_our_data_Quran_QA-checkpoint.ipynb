```python
import pandas as pd
datapath = "./new_data_less_384.csv"
df = pd.read_csv(datapath)
df.head()
```

```python
df.drop(df.columns.difference(['question','answers','passage']), 1, inplace=True)
df.head()
```

```python
# from the following repo: https://github.com/motazsaad/process-arabic-text/blob/master/clean_arabic_text.py
import re
import string
import sys
import argparse

arabic_punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''
english_punctuations = string.punctuation
punctuations_list = arabic_punctuations + english_punctuations

arabic_diacritics = re.compile("""
                             ّ    | # Tashdid
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
def remove_digits(text):
    text = re.sub(r"[1234567890١٢٣٤٥٦٧٨٩٠]+", "", text)
    return text

def remove_english_characters(text):
    text = re.sub(r'[a-zA-Z]+','',text)
    return text
    

def remove_diacritics(text):
    text = re.sub(arabic_diacritics, '', text)
    return text


def remove_punctuations(text):
    translator = str.maketrans('', '', punctuations_list)
    return text.translate(translator)
def Arabic_normalization(sentence):
    
    text = remove_english_characters(sentence)
    text = remove_digits(text)
    text = remove_diacritics(text)
    text = remove_punctuations(text)        
    return text    
```

```python
df['passage'] =df['passage'].apply(Arabic_normalization)
df.head()
```

```python
df['answers'] =df['answers'].apply(Arabic_normalization)
df.head()
```

```python
df['question'] =df['question'].apply(Arabic_normalization)
df.head()
```

```python
dataset =[]
for index , row in df.iterrows():
  #print(row)
  passage =row['passage']
  answer = row['answers']
  question= row['question']
  start_char = passage.find(answer)
  if start_char ==-1:
    print("not found !!")
    print(row)
    continue 
  #print(passage[start_char:start_char+len(answer)])
  dataset.append( { 'passage': passage, 'question':question, 'answers': [{ 'text':answer, 'start_char':start_char }]} )
```
